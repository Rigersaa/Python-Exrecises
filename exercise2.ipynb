```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\User\\Desktop\\MSc.1\\Data Vizualisation\\Python-Exrecises\\shopping_trends_updated.csv")
```

```python
df.info()
```

```python
average_purchase_by_season = df.groupby('Season')['Purchase Amount (USD)'].mean().reset_index()
plt.figure(figsize=(12,6))
bar_plot = sns.barplot(x='Purchase Amount (USD)', y='Season',data=average_purchase_by_season)
plt.xlim(55,65)
```

```python
#Grouped bars to show the purchased amount grouped by payment method and category 

group_by_payment_method = df.groupby(['Payment Method','Category'])['Purchase Amount (USD)'].mean().unstack()
```

```python
# Set style
sns.set_style('whitegrid')
# Create figure
plt.figure(figsize=(16,6))
# Plot the grouped data
group_by_payment_method.plot(kind='bar', color=sns.color_palette("Blues"))

# Adjust x-ticks and y-limit
plt.xticks(rotation=0)
plt.ylim(55, 65)

# Show the plot
plt.show()
```

```python
#Satcked bars to show the purchased amount grouped by payment method and category
stacked_bar = df.groupby(['Payment Method', 'Category'])['Purchase Amount (USD)'].mean().unstack()
stacked_bar.plot(kind='bar', stacked=True, color=sns.color_palette("Blues"))
plt.xticks(rotation=0)
plt.ylim(0, 1500)
plt.title('Average Purchase Amount by Payment Method and Category')
plt.ylabel('Average Purchase Amount (USD)')
plt.xlabel('Payment Method')
plt.show()
```
