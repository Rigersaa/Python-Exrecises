```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset (Assuming CSV file)
df = pd.read_csv('C:\\Users\\User\\Desktop\\MSc.1\\Data Vizualisation\\Python-Exrecises\\ASSIGNMENT_1_Rigersa_Bicaku\\Social_Media_Advertising.csv')
```

```python
# Data Overview
print(df.head())
print(df.info())
```

```python
print(df.columns)
```

```python
# Set color palette
sns.set_theme(style="whitegrid", palette="muted")
```

```python
# EDA: Total Impressions per Platform
plt.figure(figsize=(10, 6))
sns.barplot(x="Channel_Used", y="Impressions", data=df, estimator=sum, ci=None, palette="coolwarm")
plt.title("Total Impressions per Platform", fontsize=14, fontweight='bold')
plt.xlabel("Platform")
plt.ylabel("Total Impressions")
plt.show()
```

```python
# Grouped Bar Chart: Clicks & Engagement Score per Platform
plt.figure(figsize=(10, 6))
df_grouped = df.groupby("Channel_Used")[["Clicks", "Engagement_Score"]].sum().reset_index()
df_grouped.plot(kind="bar", x="Channel_Used", stacked=False, colormap="viridis")
plt.title("Total Clicks and Engagement Score per Platform", fontsize=14, fontweight='bold')
plt.xlabel("Platform")
plt.ylabel("Total Count")
plt.legend(["Clicks", "Engagement Score"])
plt.show()
```

```python
# 3. Stacked bar chart for Conversion Rate & Clicks
plt.figure(figsize=(8, 5))
df_pivot = df.groupby('Channel_Used')[['Clicks', 'Conversion_Rate']].sum()
df_pivot.plot(kind='bar', stacked=True)
plt.title("Conversion Rate vs Clicks per Channel")
plt.xlabel("Channel")
plt.ylabel("Total Count")
plt.show()
```

```python
# TASK 2: PERFORMANCE OVER TIME
# 4. Histogram for Clicks per day
plt.figure(figsize=(8, 5))
sns.histplot(df['Clicks'], bins=20, kde=True)
plt.title("Distribution of Clicks per Day")
plt.xlabel("Clicks")
plt.ylabel("Frequency")
plt.show()
```

```python
# TASK 3: ENGAGEMENT ANALYSIS
# 5. KDE plot for Engagement Score
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='Engagement_Score', hue='Channel_Used', fill=True)
plt.title("Density of Engagement Scores by Channel")
plt.xlabel("Engagement Score")
plt.ylabel("Density")
plt.show()
```

```python
# 6. Stacked bar chart for Impressions, Clicks, and Conversion Rate
plt.figure(figsize=(8, 5))
df_stacked = df.groupby('Channel_Used')[['Impressions', 'Clicks', 'Conversion_Rate']].sum()
df_stacked.plot(kind='bar', stacked=True)
plt.title("Proportion of Impressions, Clicks, and Conversion Rate per Channel")
plt.xlabel("Channel")
plt.ylabel("Total Count")
plt.show()
```
